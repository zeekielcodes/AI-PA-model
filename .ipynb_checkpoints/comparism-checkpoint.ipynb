```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from tensorflow.keras.models import Sequential
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from joblib import dump
```

```python
# Load your dataset
df = pd.read_csv("my-dataset.csv")
```

```python
# Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(
    df['input'], df['intent'], test_size=0.2, random_state=42
)
```

```python
# Naive Bayes
nb_model = make_pipeline(TfidfVectorizer(), MultinomialNB())
nb_model.fit(train_data, train_labels)
nb_predictions = nb_model.predict(test_data)
nb_accuracy = accuracy_score(test_labels, nb_predictions)
nb_report = classification_report(test_labels, nb_predictions, zero_division=1)

# Print accuracy and classification report
print(f"Naive Bayes Accuracy: {nb_accuracy}")
print("Naive Bayes Classification Report:\n", classification_report(test_labels, nb_predictions, zero_division=1))
```

```python
# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the training data
X_train = vectorizer.fit_transform(train_data)

# Transform the test data
X_test = vectorizer.transform(test_data)

# Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, train_labels)
rf_predictions = rf_model.predict(X_test)
rf_accuracy = accuracy_score(test_labels, rf_predictions)

rf_report = classification_report(test_labels, rf_predictions, zero_division=1)
# Print accuracy and classification report
print(f"Random Forest Accuracy: {rf_accuracy}")
print("Random Forest Classification Report:\n", classification_report(test_labels, rf_predictions, zero_division=1))
```

```python
# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the training data
X_train = vectorizer.fit_transform(train_data)

# Transform the test data
X_test = vectorizer.transform(test_data)

# K-Nearest Neighbors
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, train_labels)
knn_predictions = knn_model.predict(X_test)
knn_accuracy = accuracy_score(test_labels, knn_predictions)
knn_report = classification_report(test_labels, knn_predictions, zero_division=1)

# Print accuracy and classification report
print(f"K-Nearest Neighbors Accuracy: {knn_accuracy}")
print("K-Nearest Neighbors Classification Report:\n", knn_report)

# K-Nearest Neighbors
# knn_model = KNeighborsClassifier(n_neighbors=5)
# knn_model.fit(train_data, train_labels)
# knn_predictions = knn_model.predict(test_data)
# knn_accuracy = accuracy_score(test_labels, knn_predictions)
```

```python
#Support Vector Machine

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(train_data)
X_test = vectorizer.transform(test_data)

# Support Vector Machine (SVM) model
svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X_train, train_labels)

# Make predictions
svm_predictions = svm_model.predict(X_test)

# Evaluate the model
svm_accuracy = accuracy_score(test_labels, svm_predictions)
svm_report = classification_report(test_labels, svm_predictions, zero_division=1)
dump(svm_model, 'svm.joblib')
# Print the results
print(f"SVM Accuracy: {svm_accuracy}")
print("SVM Classification Report:\n", svm_report)
```

```python
# Convolutional Neural Network (CNN)

# Tokenize and pad sequences
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(train_data)
train_sequences = tokenizer.texts_to_sequences(train_data)
max_len = max(len(seq) for seq in train_sequences)
train_sequences_padded = pad_sequences(train_sequences, maxlen=max_len)

# Label encoding
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_labels)

# Convolutional Neural Network (CNN) model
cnn_model = Sequential()
cnn_model.add(Embedding(input_dim=5000, output_dim=50, input_length=max_len))
cnn_model.add(Conv1D(128, 5, activation='relu'))
cnn_model.add(GlobalMaxPooling1D())
cnn_model.add(Dense(64, activation='relu'))
cnn_model.add(Dense(1, activation='sigmoid'))

# Split data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    train_sequences_padded, train_labels_encoded, test_size=0.2, random_state=42
)


# Compile the model
cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
cnn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_val, y_val))

# Tokenize and pad sequences for test data
test_sequences = tokenizer.texts_to_sequences(test_data)
test_sequences_padded = pad_sequences(test_sequences, maxlen=max_len)

# Predictions on test data
cnn_predictions = cnn_model.predict(test_sequences_padded)

# Convert probabilities to class labels using a threshold (e.g., 0.5)
predicted_labels_cnn = (cnn_predictions > 0.5).astype(int)

# Inverse transform labels to original classes
predicted_labels_original_cnn = label_encoder.inverse_transform(predicted_labels_cnn.flatten())

# Print the predictions
print(predicted_labels_original_cnn)

# Calculate accuracy
cnn_accuracy = accuracy_score(test_labels, predicted_labels_original_cnn)

# Generate classification report
cnn_report = classification_report(test_labels, predicted_labels_original_cnn, zero_division=1)

# Print accuracy and classification report
print(f"CNN Accuracy: {cnn_accuracy}")
print("CNN Classification Report:\n", cnn_report)
```
